```python
import spacy
from chatterbot.trainers import ListTrainer
from chatterbot import ChatBot # Changed chatbot to ChatBot
nlp = spacy.load("en_core_web_sm")
nlp = spacy.blank("en")
chatbot = ChatBot("New_Bot") # Changed chatbot to ChatBot
```

```python
conversation = ["Hello",
"Hi, there!",
"How are you doing?",
"That is good to hear",
"How can you I help you",
"Thank you",
"You’re welcome"
]
```

```python
trainer = ListTrainer(chatbot)
trainer.train(conversation)
print("Enter 'quit' to stop")
while True:
  text_input = input ("You:")
  if text_input == 'quit':
    break
  print("Bot:", chatbot.get_response(text_input))
```
